```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
trail_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_trial_df = pd.merge(mouse_df, trail_df, how='outer')
combined_trial_df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_time_series = combined_trial_df.groupby(['Drug','Timepoint'])
tumor_volume_mean=drug_time_series.mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
tumor_volume_df = pd.DataFrame(tumor_volume_mean)

# Preview DataFrame
tumor_volume_df_preview=tumor_volume_df.reset_index()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

standard_errors = drug_time_series.sem()['Tumor Volume (mm3)']
standard_errors

# Convert to DataFrame
sem_df = pd.DataFrame(standard_errors)
sem_df

# Preview DataFrame
sem_df_preview=sem_df.reset_index()
sem_df_preview
```

```python
reformat_error_df = pd.pivot_table(sem_df, index ='Timepoint', columns ='Drug', values="Tumor Volume (mm3)")
reformat_error_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_df = pd.pivot_table(tumor_volume_df, index ='Timepoint', columns ='Drug', values="Tumor Volume (mm3)")
reformat_df
```

```python
# Generate the Plot (with Error Bars)

Capomulin, = plt.plot(reformat_df["Capomulin"],color="red", label="Capomulin", marker="o",linestyle = "--")
Infubinol, = plt.plot(reformat_df["Infubinol"], color="blue", label="Infubinol", marker="^", linestyle = "--" )
Ketapril, = plt.plot(reformat_df["Ketapril"], color="green", label="Ketapril", marker="s", linestyle = "--")
Placebo, = plt.plot(reformat_df["Placebo"], color="black", label="Placebo", marker="d", linestyle = "--")


plt.errorbar(reformat_df["Capomulin"].index, reformat_df["Capomulin"], yerr=reformat_error_df["Capomulin"], fmt='o', color='red')
plt.errorbar(reformat_df["Infubinol"].index, reformat_df["Infubinol"], yerr=reformat_error_df["Infubinol"], fmt='o', color='blue')
plt.errorbar(reformat_df["Ketapril"].index, reformat_df["Ketapril"], yerr=reformat_error_df["Ketapril"], fmt='o', color='green')
plt.errorbar(reformat_df["Placebo"].index, reformat_df["Placebo"], yerr=reformat_error_df["Placebo"], fmt='o', color='black')


plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Time Response to Treatment")
plt.grid(axis="y")
plt.legend(loc="upper left")
# Save the Figure

#plt.savefig(os.path.join('Images','tumor_response_to_treatment.png'))
plt.savefig("../Images/Tumor Response To Treatment.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_site_series = combined_trial_df.groupby(['Drug','Timepoint'])
met_site_mean=met_site_series.mean()["Metastatic Sites"]

# Convert to DataFrame
met_site_df = pd.DataFrame(met_site_mean)

# Preview DataFrame
met_site_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_standard_errors = met_site_series.sem()["Metastatic Sites"]
met_standard_errors

# Convert to DataFrame
met_sem_df = pd.DataFrame(met_standard_errors)
met_sem_df

met_error_df = pd.pivot_table(met_sem_df, index ='Timepoint', columns ='Drug', values="Metastatic Sites")
met_error_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_met_df = pd.pivot_table(met_site_df, index ='Timepoint', columns ='Drug', values="Metastatic Sites")
reformat_met_df
```

```python
# Generate the Plot (with Error Bars)
Capomulin, = plt.plot(reformat_met_df["Capomulin"],color="red", label="Capomulin", marker="o",linestyle = "--")
Infubinol, = plt.plot(reformat_met_df["Infubinol"], color="blue", label="Infubinol", marker="^", linestyle = "--" )
Ketapril, = plt.plot(reformat_met_df["Ketapril"], color="green", label="Ketapril", marker="s", linestyle = "--")
Placebo, = plt.plot(reformat_met_df["Placebo"], color="black", label="Placebo", marker="d", linestyle = "--")


plt.errorbar(reformat_met_df["Capomulin"].index, reformat_met_df["Capomulin"], yerr=met_error_df["Capomulin"], fmt='o', color='red')
plt.errorbar(reformat_met_df["Infubinol"].index, reformat_met_df["Infubinol"], yerr=met_error_df["Infubinol"], fmt='o', color='blue')
plt.errorbar(reformat_met_df["Ketapril"].index, reformat_met_df["Ketapril"], yerr=met_error_df["Ketapril"], fmt='o', color='green')
plt.errorbar(reformat_met_df["Placebo"].index, reformat_met_df["Placebo"], yerr=met_error_df["Placebo"], fmt='o', color='black')


plt.xlabel("Treatment Duration(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Metastatic Spread During Treatment")
plt.grid(axis="y")
plt.legend(loc="upper left")
# Save the Figure
plt.savefig("../Images/Metastic Spread During Treatment.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count_series = combined_trial_df.groupby(['Drug','Timepoint'])
mouse_count=mouse_count_series.nunique()["Mouse ID"]

# Convert to DataFrame
mouse_count_df = pd.DataFrame(mouse_count).rename({"Mouse ID": "Mouse Count"})

# Preview DataFrame
mouse_count_df_preview=mouse_count_df.reset_index()
mouse_count_df_preview
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_mouse_count_df = pd.pivot_table(mouse_count_df, index ='Timepoint', columns ='Drug', values="Mouse ID")
reformat_mouse_count_df
```

```python
# Generate the Plot (Accounting for percentages)

Capomulin, = plt.plot((reformat_mouse_count_df["Capomulin"]/reformat_mouse_count_df["Capomulin"][0])*100,color="red", label="Capomulin", marker="o",linestyle = "--")
Infubinol, = plt.plot((reformat_mouse_count_df["Infubinol"]/reformat_mouse_count_df["Infubinol"][0])*100, color="blue", label="Infubinol", marker="^", linestyle = "--" )
Ketapril, = plt.plot((reformat_mouse_count_df["Ketapril"]/reformat_mouse_count_df["Ketapril"][0])*100, color="green", label="Ketapril", marker="s", linestyle = "--")
Placebo, = plt.plot((reformat_mouse_count_df["Placebo"]/reformat_mouse_count_df["Placebo"][0])*100, color="black", label="Placebo", marker="d", linestyle = "--")


plt.xlabel("Time(Days)")
plt.ylabel("Survival Rate(%)")
plt.title("Survival During Treatment")
plt.grid()
plt.legend(loc="upper left")
# Save the Figure
plt.savefig("../Images/Survival During Treatment")
# Show the Figure
plt.show()
```
